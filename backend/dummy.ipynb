```python
import random
import time
import json
```

```python


# List of 10 random stock symbols
stock_symbols = ["AAPL", "MSFT", "GOOGL", "AMZN", "TSLA", "NFLX", "NVDA", "META", "ADBE", "INTC"]

# Function to generate random stock data
def generate_stock_data(num_records=50):
    """Generate random stock data for 10 stocks."""
    stock_data = {symbol: [] for symbol in stock_symbols}
    
    for _ in range(num_records):
        for symbol in stock_symbols:
            # Generate a random price and timestamp
            price = round(random.uniform(100, 1000), 2)  # Random price between $100 and $1000
            timestamp = time.strftime("%Y-%m-%d %H:%M:%S")
            
            # Append the data to the stock's list
            stock_data[symbol].append({"timestamp": timestamp, "price": price})
    return stock_data
        # Pause to simulate real-time data generation
```

```python
data = generate_stock_data()
```

```python
with open("stock_price.json", "w") as f:
    json.dump(data, f, indent=4) 
```
